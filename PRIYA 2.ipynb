```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import LabelEncoder
```

```python
df =pd.read_csv("T:/GIT HUB PROJECTS/CUSTOMER CURN/Churn_Modelling.csv")
```

```python
label_encode = LabelEncoder()
```

```python
label = label_encode.fit_transform(df.Gender)
```

```python
df['Gender']= label
```

```python
df
```

```python
label2= label_encode.fit_transform(df.Geography)
```

```python
df['Geography']=label2
```

```python
df
```

```python
x=df.drop(['Surname','Exited'],axis=1)
```

```python
y=df['Exited']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=35)
```

```python
x_train.shape
```

```python
linear_model=LogisticRegression()
```

```python
linear_model.fit(x_train,y_train)
```

```python
y_predict= linear_model.predict(x_test)
```

```python
accuracy= accuracy_score(y_test,y_predict)
```

```python
accuracy
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rtf=RandomForestClassifier()
```

```python
rtf.fit(x_train,y_train)
```

```python
y_predict =rtf.predict(x_test)
```

```python
accuracy=accuracy_score(y_predict,y_test)
```
